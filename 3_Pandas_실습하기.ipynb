```python
# 수학점수 내림차순
df_exam.sort_values(by='math', ascending=False)
```

```python
# 영어 점수 60~ 70, 내림차순
df_exam[(df_exam.english >=60) & (df_exam.english <= 70)].sort_values(by='math', ascending=False)
```

```python
import math
import numpy as np

# 파씽변수
df_exam['total'] = df_exam.math + df_exam.english + df_exam.science
df_exam['mean'] = round(df_exam.total / 3)
# 'A' if df_exam.mean >= 90 else 'B'
df_exam['grade'] = np.where(df_exam['mean'] >= 90, 'A', np.where(df_exam['mean'] >= 80, 'B', np.where(df_exam['mean'] >= 70, 'C', np.where(df_exam['mean'] >= 60, 'D', 'F'))))
df_exam
```
