```python
"""
날짜 : 2021/07/05
이름 : 이진유
내용 : 파이썬 Matplotlib 실습하기
"""

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 기본 플롯차트
plt.title('plot1')
plt.plot([1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# x축 설정
plt.title('plot2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# 차트 스타일
plt.title('plot3')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17], 'go:') # g : green, o :동그라미 ':': 점선, '-' : 실선 '--' : 대쉬선 ->  예) 'bs:' 파란색 네모모양 점선 그래프
plt.grid()
plt.show()
```

```python
# 멀티 라인차트
values = np.arange(0,5, 0.2)
print(values)

plt.title('plot4')
plt.plot(values, values, 'r--') # g : green, o :동그라미 ':': 점선, '-' : 실선 '--' : 대쉬선 ->  예) 'bs:' 파란색 네모모양 점선 그래프
plt.plot(values, values ** 2, 'g--')
plt.plot(values, values ** 3, 'b--')
plt.grid()
plt.grid()
plt.show()
```

```python
# 막대 그래프
plt.title('Bar chart')
plt.bar(['apple','barnana','orange','grape','berry'],[5,3,7,9,8])
plt.xlabel('fruit')
plt.ylabel('count')
plt.show()
```

```python
# 히스토그램
data1 = np.random.rand(1000) # 0 ~ 1 사이의 임의의 실수 1000개
data2 = np.random.randn(1000) # 0 ~ 1 사이의 정규분포를 따라는 임의 실수 1000개

plt.title('Histogram')
plt.hist(data1, edgecolor='black') # edgecolor 테두리
plt.show()
```

```python
plt.title('Histogram2')
plt.hist(data2, edgecolor='black') # edgecolor 테두리
plt.show()
```

```python
# 산점도
x = np.random.normal(0, 1, 1000) # 표준 편차가 1인 정규분포
y = np.random.normal(0, 1, 1000)

plt.title('scatter')
plt.scatter(x, y)
plt.show()
```
