```python
#한글 형태소 분석기 생성
from konlpy.tag import Okt
okt = Okt()
```

```python
#형태소 분석
result1 = okt.pos('아버지가 방에 들어 가십니다.')
print('result1 : ', result1)

result2 = okt.nouns('나는 어제 보리밥을 먹었다.')
print('result2 : ', result2)
```

```python
#텍스트 파일 불러오기
sample = open('/content/drive/MyDrive/Colab Notebooks/file/wordcloud_data.txt').read()
sample

sample_nouns = okt.nouns(sample)
sample_nouns = ' '.join(sample_nouns) # List에 있는 데이터를 한칸씩 띄워서 문자열로 연결

#단어구름 출력
wc = WordCloud(width=1200,
               height=1200,
               font_path='/content/drive/MyDrive/Colab Notebooks/file/malgun.ttf',
               background_color='white').generate(sample_nouns)
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
